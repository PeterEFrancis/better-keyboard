```python
import random
```

```python
NUM_TRIALS = 1000

# upper bound calculations

upper_bounds = {}



def num_trials_met(percent, pop_size):
    num_met = 0
    for i in range(NUM_TRIALS):
        samples = set()
        for _ in range(int(percent * pop_size)):
            samples.add(random.randint(1, pop_size))
        if len(samples) == pop_size:
            num_met += 1
    return num_met / NUM_TRIALS


def find_50_confidence(pop_size):
    # find upper bound on percent_needed
    upper_bound = 1
    while True:
        if num_trials_met(upper_bound, pop_size) < 0.5:
            upper_bound *= 10
        else:
            break
    # find actual estimate using binary search
    lower_bound = 1
    while True:
        avg = (lower_bound + upper_bound) / 2
        num = num_trials_met(avg, pop_size)
        if abs(num - 0.5) < 0.1:
            return avg
        if num < 0.5:
            lower_bound = avg
        else:
            upper_bound = avg
```

```python
data = []
for i in range(2, 200):
    c = find_50_confidence(i)
    data.append(c)
    print(c)
```
